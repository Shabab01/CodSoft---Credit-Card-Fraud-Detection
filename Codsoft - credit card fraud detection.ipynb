```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

Preprocessing

Splitting the data into training and testing sets

```python
X = data.drop('Class', axis=1)  # Features
y = data['Class']  # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Training the logistic regression model

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

Predicting on the test set

```python
y_pred = model.predict(X_test)
```

Evaluating the model

```python
a1 = accuracy_score(y_test, y_pred)
print("Accuracy:", a1)

# Calculate Precision
p1 = precision_score(y_test, y_pred)
print("Precision:", p1)

# Calculate Recall
r1 = recall_score(y_test, y_pred)
print("Recall:", r1)

# Calculate F1 Score
f1_1 = f1_score(y_test, y_pred)
print("F1 Score:", f1_1)

c1 = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", c1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

```python
y_pred = dtree.predict(X_test)
```

Evaluating the model

```python
a4 = accuracy_score(y_test, y_pred)
print("Accuracy:", a4)

# Calculate Precision
p4 = precision_score(y_test, y_pred)
print("Precision:", p4)

# Calculate Recall
r4 = recall_score(y_test, y_pred)
print("Recall:", r4)

# Calculate F1 Score
f1_4 = f1_score(y_test, y_pred)
print("F1 Score:", f1_4)

c4 = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", c4)
```
